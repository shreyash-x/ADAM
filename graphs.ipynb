```python
import numpy as np
import csv
import matplotlib.pyplot as plt
def plot_loss(file_name,title):
    plt.figure(figsize=(10, 10))
    with open(file_name) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        for row in csv_reader:
            optimizer_name = row[0]
            loss_values = row[1:]
            loss_values = [float(i) for i in loss_values]
            x_axis = []
            for i in range(len(loss_values)):
                x_axis.append(i+1)
            plt.plot(x_axis,loss_values,label=optimizer_name)
    # size of the figure
    plt.title(title)
    plt.xlabel("epochs")
    plt.ylabel("loss")
    plt.legend()
    plt.show()
```

```python
plot_loss("nn_lr2.csv","Neural Network with learning rate 0.01")
plot_loss("nn_lr3.csv","Neural Network with learning rate 0.001")
```

```python
plot_loss("cnn_lr2.csv","Convolutional Neural Network with learning rate 0.01")
plot_loss("cnn_lr3.csv","Convolutional Neural Network with learning rate 0.001")
```

```python
plot_loss("lr.csv","Logisitic Regression with learning rate 0.001")
```
